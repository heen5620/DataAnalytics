```python
import folium
```

```python
df_Alzheimer = df[df["상권업종소분류명"].str.contains("노인" or "치매")]
```

```python
lat = df_Alzheimer["위도"].mean()
lat
```

```python
long = df_Alzheimer["경도"].mean()
long
```

```python
m = folium.Map(location=[lat,long], zoom_start=10)
m
```

```python
for index, row in df_Alzheimer.iterrows():
    name = row["상호명"]
    address = row["도로명주소"]
    popup = f"{name}-{address}"
    location = [lat,long]
    
    folium.Marker(
        location=location,
        popup=popup
    ).add_to(m)
```
